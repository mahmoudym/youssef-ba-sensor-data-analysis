```python
#Dependencies
import pandas as pd 
import numpy as np
np.random.seed(550)
import math
from keras.models import Sequential
from keras import optimizers
from keras.layers import LSTM
from keras.layers import Dense, Activation
import keras 
from keras.metrics import categorical_accuracy
```

```python
data = pd.read_csv("finalizeddata.csv",sep=",",header=None)
```

```python
#A method to take each 2 rows and combine them into 1 row to pass them to the RNN later
def configure_dataframe(data):
    result = pd.DataFrame(np.random.randint(low=0, high=1, size=(int(data.shape[0]/2), data.shape[1])))
    j=0 #to manage iterating over the data(taking tuples into consideration)
    
    #To change the datatypes of each cell so it can support tuples (TODO:could be done in a cleaner way )
    for column in result:
            result[column]=result[column].astype(object)

    # Main iteration 
    for index,row in data.iterrows():
        #Setting the first 3 Columns
           if(j%2==0):
                result[0][int(index/2)]= data.iloc[index][0]
                result[1][int(index/2)]=data.iloc[index][1]
                result[2][int(index/2)]=data.iloc[index][2]
           
        
           for i in range(3,len(row)):
                if(j%2==0):
                    #If it is an odd row. just take the value and put it in the cell
                    result[i][int(index/2)]=data.iloc[index][i]
                else:
                    #If it is an even row. combine the data with the existing row to turn them to tuple 
                    result[i][int(index/2)]= [result[i][int(index/2)],data.iloc[index][i]]
                    
           j+=1
    return result
```

```python
def further_configuration(data):
    result = np.zeros(shape=(data.shape[0],data.shape[1],2))
    for index,row in data.iterrows():
            for i in range(0,len(row)):
                #i+3 because we deleted the first 3 rows of the dataset 
                result[index][i][0]=data[i+3][index][0]
                result[index][i][1]=data[i+3][index][1]
          
    return result
```

```python
# 1 => [1,0]  // 2 => [0,1] easier for the algorithm to predict
def one_hot_output(data):
    output = pd.DataFrame(np.random.randint(low=0, high=1, size=(data.shape[0], 2)))
    i=0
    for row in data:
        output[int(row)-1][i]=1
        i+=1
    
    return output
            
```

```python
data2 = configure_dataframe(data)
```

```python
train_output = one_hot_output(data2[0])
train_input= data2.copy()

train_input.drop(train_input.columns[0:3], axis=1, inplace=True)
train_input = further_configuration(train_input)
```

```python
def lstm(NUM_EXAMPLES):
    global train_input
    global train_output
    t_input = train_input
    t_output = train_output
    test_input = t_input[NUM_EXAMPLES:]
    test_output = t_output[NUM_EXAMPLES:] #everything beyond num examples
    t_input = t_input[:NUM_EXAMPLES]
    t_output = t_output[:NUM_EXAMPLES] #till num examples
    print(test_input.shape)
    print(t_input.shape)
    model = Sequential()
    model.add(LSTM(units=32, activation="sigmoid", input_shape=(train_input.shape[1],train_input.shape[2])))
    model.add(Dense(16,activation='sigmoid'))
    model.add(Dense(2,activation='sigmoid'))


    adam1 = keras.optimizers.Adamax(lr=0.005157, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0)
    #model.compile(loss='categorical_crossentropy',optimizer='adam',
                # metrics=[categorical_accuracy])
    model.compile(loss='categorical_crossentropy', optimizer=adam1,metrics=[categorical_accuracy])



    model.fit(t_input, t_output, epochs=210,verbose=2,steps_per_epoch = 10)

    score, acc = model.evaluate(test_input, test_output)
    print ('test score:',score)
    print ('test acc:', acc)
    print (model.predict_classes(test_input))
    print (test_output)
```
